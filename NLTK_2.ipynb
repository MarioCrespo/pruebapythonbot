En Python, **encoding='utf8'** es un parámetro opcional que se puede pasar a la función open() para especificar la codificación de caracteres que se utilizará para leer el archivo de texto.

La codificación de caracteres es el proceso de transformar los caracteres de un archivo de texto en una secuencia de bytes para que puedan ser almacenados en el disco o transmitidos a través de una red. La codificación de caracteres determina cómo se representarán los caracteres en la memoria de la computadora y cómo se decodificarán posteriormente para su visualización o procesamiento.

utf-8 es una codificación de caracteres que puede representar cualquier carácter Unicode, lo que la hace muy útil para trabajar con texto en diferentes idiomas. Por lo tanto, si el archivo de texto que se va a leer contiene caracteres no-ASCII (por ejemplo, letras con acentos, caracteres japoneses, etc.), es importante especificar la codificación de caracteres correcta para evitar problemas de lectura y escritura.

encoding='utf8' en f = open('/content/Quijote.txt', encoding='utf8') especifica que el archivo de texto "Quijote.txt" se leerá utilizando la codificación de caracteres utf-8, lo que garantiza que se puedan leer correctamente los caracteres no-ASCII que pueda contener el archivo.

**En informática, un carácter Unicode** es una representación estándar y universal de un carácter, independientemente del lenguaje o del sistema de escritura utilizado. Unicode es un estándar de codificación de caracteres que asigna un número único a cada carácter, símbolo o signo de puntuación que se utiliza en la mayoría de los sistemas de escritura conocidos en todo el mundo.

Si queremos usarlo con las funciones del nltk, lo tenemos que integrar en él. Pues se carga como lista, por lo que tenemos que pasarlo como una lista de elementos. Por ejemplo, split(), ¿os acordáis? ¿Qué hacía eso en las str ?

```python
import nltk
text_quijote_en_nltk = nltk.Text(texto_plano_cargado.split())#finalmente se carga en nltk
```

```python
print(text_quijote_en_nltk)
```

```python
len(text_quijote_en_nltk)
```

```python
text_quijote_en_nltk.common_contexts(["computer","language"])
```
